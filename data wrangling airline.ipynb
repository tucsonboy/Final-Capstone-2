```python
import pandas as pd
import os
import glob
```

```python
# Data was downloaded by month and year. Going to combine all the files to create one data frame.
df = []
all_files = glob.glob('c:/calvin/*.csv')


```

```python
for filename in all_files:
    temp = pd.read_csv(filename, encoding = 'latin-1')
    df.append(temp)
```

```python
data = pd.concat(df, axis = 0)
data.info()
```

```python
# Check for NaN in the data
data.isnull().sum()
```

```python
data.head()
```

```python
# When there isn't any delays, the dataframe has a value of NaN,
#I am going to convert those to 0 in columns dealing with delays
delays = ['CARRIER_DELAY','WEATHER_DELAY', 'NAS_DELAY', 'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY']
for column in delays:
    data[column] = data[column].fillna(0)
data.head()    
```

```python
# Flights that are going in and out of MIA (Miami International Airport) are needed.
# All other flights will be removed.

data1 = data.query('ORIGIN == "MIA" or DEST == "MIA"')
print(data1['ORIGIN'])
print(data1['DEST'])
```

```python
#Save wrangled data frame to csv
data1.to_csv('c:/calvin/mia_data.csv')
```
